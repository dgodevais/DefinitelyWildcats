```python
# Load the jpg file into a numpy array
file_path = "/Users/hollisnymark/Documents/Grad School/Course Work/2018 Spring/Big Data/Group Project/wiki_crop/00/37500_1944-01-23_2010.jpg"
image = face_recognition.load_image_file(file_path)


# Find all the faces in the image using the default HOG-based model.
# This method is fairly accurate, but not as accurate as the CNN model and not GPU accelerated.
# See also: find_faces_in_picture_cnn.py
face_locations = face_recognition.face_locations(image)


    #PRINTS the whether or not a face was found in the photo -- not necessary   
#print("I found {} face(s) in this photograph.".format(len(face_locations)))

for face_location in face_locations:
    
    #set face locations
    top, right, bottom, left = face_location
        #PRINTS the location of each face in this image
    #print("A face is located at pixel location Top: {}, Left: {}, Bottom: {}, Right: {}".format(top, left, bottom, right))

    # You can access the actual face itself like this:
    face_image = image[top:bottom, left:right]
    pil_image = Image.fromarray(face_image)


    
#convert cropped image to greyscale
c_pil_image = pil_image.convert('LA')


#resize cropped image
size = 500,500
final = c_pil_image.resize(size)
final
    
```

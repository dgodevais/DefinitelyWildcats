```python
import numpy as np
import pandas as pd

from PIL import Image
```

```python
#convert cropped image to greyscale
# Load the jpg file into a numpy array
file_path = "/Users/hollisnymark/Documents/Grad School/Course Work/2018 Spring/Big Data/Group Project/wiki_crop/00/37500_1944-01-23_2010.jpg"
image = Image.open(file_path)

#conver to black and white
c_pil_image = image.convert('LA')

#resize cropped image
size = 500,500
final = c_pil_image.resize(size)

#change image to numpy array
pix = np.array(final)

#create 2-d numpy array with white values only (black value = 255)

rows = []
for i in range(0,500):
    row = pix[i] 
    pixels = []
    for a in row:
        pixel = row[i][0]
        pixels.append(pixel) 
    rows.append(pixels)

rows = np.array(rows)

#create dataframe from white values -- pixel location are headers/columns
df = pd.DataFrame(data=rows)
df

```
